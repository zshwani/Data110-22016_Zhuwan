```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
# Pair Plot
sns.pairplot(iris, hue='species', corner=True)
plt.suptitle('Pair Plot of Iris Features by Species', y=1.02)
plt.show()
```

```python
# Bar Plot for average sepal length
sns.barplot(x='species', y='sepal_length', data=iris, ci=None)
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.show()
```

```python
# Box Plot for petal width
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title('Distribution of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# Grouped Bar Plot for average sepal and petal lengths
grouped_data = iris.groupby('species')[['sepal_length', 'petal_length']].mean().reset_index()
melted_data = grouped_data.melt(id_vars='species', var_name='Feature', value_name='Length')

sns.barplot(x='species', y='Length', hue='Feature', data=melted_data, ci=None)
plt.title('Average Sepal and Petal Lengths by Species')
plt.xlabel('Species')
plt.ylabel('Average Length (cm)')
plt.show()
```

```python
# Violin Plot for sepal length
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# Aggregation of numerical features
aggregation = iris.groupby('species').agg(['mean', 'min', 'max'])
print(aggregation)
```
